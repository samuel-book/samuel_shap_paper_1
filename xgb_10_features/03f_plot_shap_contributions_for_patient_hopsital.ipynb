# Plotting the contribution of patient, hospital, and patient-hospital interactions to thrombolysis decision-making

Data from notebook 3e.
Data comes from the sum of absolute SHAP values.

```python
import matplotlib.pyplot as plt
```

```python
sum_abs_shap = {
    'patient': 444218,
    'hospital': 139988,
    'hospital-patient interaction': 88098}

values = sum_abs_shap.values()

```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()
ax.pie(sum_abs_shap.values(),
      labels=sum_abs_shap.keys(),
      autopct='%1.0f%%')

# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle( (0,0), 0.3, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)


plt.savefig('./output/3f_shap_contributions.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
