# An example showing how hopsital SHAP can have a more pronounced effect on strokes with low stroke severity

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Hospital SHAP = -1 
df1 = pd.DataFrame()
base_probs = 0.5
df1['base prob'] = np.repeat(base_probs, 6)
df1['base odds'] = df1['base prob'] / (1 - df1['base prob'])
df1['base log odds'] = np.log(df1['base odds'] )
df1['nihss'] = [0, 5, 10, 15, 20, 25]
df1['nihss shap'] = [-2.5, -1, 0.8, 1, 1, 1]
df1['arrival to scan shap'] = 1.5
df1['hospital shap'] = -1
df1['base log odds']
df1['total shap'] = df1['nihss shap'] + df1['hospital shap'] + df1['arrival to scan shap']
df1['adjusted log odds'] = df1['base log odds'] + df1['total shap']
df1['adjusted odds'] = np.exp(df1['adjusted log odds'])
df1['adjusted probability'] = df1['adjusted odds'] / (df1['adjusted odds'] + 1)
df1
```

```python
# Hospital SHAP = 0
df2 = pd.DataFrame()
base_probs = 0.5
df2['base prob'] = np.repeat(base_probs, 6)
df2['base odds'] = df2['base prob'] / (1 - df2['base prob'])
df2['base log odds'] = np.log(df2['base odds'] )
df2['nihss'] = [0, 5, 10, 15, 20, 25]
df2['nihss shap'] = [-2.5, -1, 0.8, 1, 1, 1]
df2['arrival to scan shap'] = 1.5
df2['hospital shap'] = 0
df2['base log odds']
df2['total shap'] = df2['nihss shap'] + df2['hospital shap'] + df2['arrival to scan shap']
df2['adjusted log odds'] = df2['base log odds'] + df2['total shap']
df2['adjusted odds'] = np.exp(df2['adjusted log odds'])
df2['adjusted probability'] = df2['adjusted odds'] / (df2['adjusted odds'] + 1)
df2
```

```python
# Hospital SHAP = +1
df3 = pd.DataFrame()
base_probs = 0.5
df3['base prob'] = np.repeat(base_probs, 6)
df3['base odds'] = df3['base prob'] / (1 - df3['base prob'])
df3['base log odds'] = np.log(df3['base odds'] )
df3['nihss'] = [0, 5, 10, 15, 20, 25]
df3['nihss shap'] = [-2.5, -1, 0.8, 1, 1, 1]
df3['arrival to scan shap'] = 1.5
df3['hospital shap'] = 1
df3['base log odds']
df3['total shap'] = df3['nihss shap'] + df3['hospital shap'] + df3['arrival to scan shap']
df3['adjusted log odds'] = df3['base log odds'] + df3['total shap']
df3['adjusted odds'] = np.exp(df3['adjusted log odds'])
df3['adjusted probability'] = df3['adjusted odds'] / (df3['adjusted odds'] + 1)
df3
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(df1['nihss'], df1['adjusted probability'],  ls='-', marker='o', c='b', label='Hopsital SHAP = -1')
ax.plot(df2['nihss'], df2['adjusted probability'],  ls='-', marker='s', c='r', label='Hopsital SHAP = 0')
ax.plot(df3['nihss'], df3['adjusted probability'],  ls='-', marker='^', c='y', label='Hopsital SHAP = 1')
ax.legend()
ax.set_xlabel('Stroke severity (NIHSS)')
ax.set_ylabel('Probability of receiving thrombolysis')
ax.set_ylim(0,1)
ax.set_xlim(-0.5,25.5)
ax.grid()
ax.plot()
plt.show()
```
